```python
seasons_stats = seasons_stats.fillna(0)
seasons_stats = seasons_stats[seasons_stats.Tm != 'TOT']
```

```python
# Function to create 'xGame' column, to calculate the average for stats per game
# x == DataFrame ; y == Column

def pergame(x, y):
    x[y + 'xG'] = round(x[y]/x['G'], 2)
```

```python
fig, ax = plt.subplots(figsize=(14, 10))
sns.distplot(seasons_stats['Age']);

plt.title("Age Distribution")
plt.show()
```

```python
seasons_stats['ppg'] = seasons_stats.PTS/seasons_stats.G
```

```python
ppg = pd.DataFrame()
ppg['player'] = seasons_stats.groupby('Player').mean()['ppg'].index
ppg['ppg'] = seasons_stats.groupby('Player').mean()['ppg'].values
ppg['Games played'] = seasons_stats.groupby('Player').sum()['G'].values
ppg['points'] = seasons_stats.groupby('Player').mean()['PTS'].values
ppg = ppg.sort_values('ppg', ascending=False).head(10)
ppg
```
